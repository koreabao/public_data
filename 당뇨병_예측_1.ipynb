```python
data=pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv',names=['임신횟수', '공복혈당', '낮은혈압', '삼두근주름', '혈청인슐린', '체질량BMI', '당뇨가족력','나이', '정상0당뇨1'])
# 데이터 변수 영문명 names=['pregnant','plasma', 'pressure','thickness','insulin','BMI','pedigree','age','diagnosis'])
# 데이터 변수 한글명 names=['임신횟수', '공복혈당', '낮은혈압', '삼두근주름', '혈청인슐린', '체질량BMI', '당뇨가족력','나이', '정상0당뇨1'])
```

```python
x=data.iloc[:,:-1]
x
```

```python
y = data.iloc[:,[-1]]
y
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='lbfgs',max_iter=2000)
```

```python
model.fit(x,y)
```

```python
model.predict(x)
```

```python
# 실제값과 예측값 비교 
for i in range(10):                                              # for i in range(len(x)):
    label =  y.iloc[i, -1]                                         # DataFrame.iat[index_int, column_int]   #.iloc와 동일하게 위치 정수로 데이터에 접근하는 메소드임   ### DataFrame.at['index_label', 'column_label']  # .at은 .loc와 동일하게 레이블label기반으로 데이터 접근
    prediction = model.predict(x)[i]
    print('실제: {:.0f}, 예측: {:.0f}'.format(prediction, label))
```

```python
# 모델 설명력 확인(R2) (3) : accuracy_score, Machine Learning, Deep Learning 분석에서 사용
from sklearn.metrics import accuracy_score
y_pred = model.predict(x)
print('모델설명력: ', accuracy_score(y_pred, y))
```

```python
# 독립변수 각각의 영향력과 유의성 확인: 회귀계수베타의 유의수준(P)으로 확인함

# variable_setting = 'diagnosis ~ plasma + pressure + thickness + insulin + BMI + pedigree + age'                                                      # 종속변수 ~ 독립변수들
variable_setting = '정상0당뇨1 ~ 임신횟수 + 공복혈당 + 낮은혈압 + 삼두근주름 + 혈청인슐린 + 체질량BMI + 당뇨가족력 + 나이'

import statsmodels.api as sm
logit = sm.Logit.from_formula(variable_setting, data).fit()    # from_formula('종속변수 ~ 독립변수들', 파일명)

print(logit.summary())         # P>|t|(유의확률) :  독립변수의 유의확률, 0.05보다 작을 때 독립변수가 종속변수에 미치는 영향이 유의하다고 평가함
                                                 # 유의수준(有意水準 significance level): 수학 통계적 검정에서 그 가설이 옳은데도 불구하고 틀린 것으로 치고 기각하는 확률의 정밀도를 나타내는 수.  보통 1%, 5%, 즉,0.01, 0.05를 주로 사용함 
                                                 # 유의확률(有意 確率 significance probability, p-value): 유의수준을 보다 정확히 계산한 것
                                                 # 통계적 유의성: 어떤 가설이 단순한 우연이 아니라 의미를 갖는다는 것을 의미함
                                                 # 유의(留意): 마음에 새겨 두어 조심하며 관심을 가짐의 의미. 주의(注意)와 비슷한 말
                                              
```
